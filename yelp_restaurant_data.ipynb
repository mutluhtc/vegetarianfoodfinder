```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

# Yelp Restaurant Menu Datasets

```python
cities = ['berkeley','west_lafayette','st_louis','san_francisco','salt_lake_city',
          'seattle','santa_cruz_ca','portland','phoenix','miami','washington_dc',
          'atlanta','austin','boston','chicago','dallas','denver_co','detroit','honolulu',
          'houston','indianapolis','las_vegas','los_angeles_ca','minneapolis_mn','new_york_city',
          'orlando_fl','philadelphia_pa','san_diego','san_jose']

cities.sort() # Sort cities alphabetically (directly modifies list)

city_df_list = []

for name in cities:
    file_path_unix = f"structured_JSON_data_files/{name}_restaurants_with_structured_menus.json"
    df = pd.read_json(file_path_unix)
    city_df_list.append(df)

# Merge all downloaded dataframes into one
yelp_full_df = pd.concat(city_df_list)
yelp_full_df.reset_index(drop=True, inplace=True)

print(yelp_full_df.shape)
yelp_full_df.head()
```

```python
yelp_full_df.tail()
```

```python
# Filter and keep all rows that have a nonempty menu

nonempty_list_rows = yelp_full_df['menu'].apply(lambda x: x!=[])

# New dataframe with nonempty menus
yelp_menu_df = yelp_full_df.loc[nonempty_list_rows]
yelp_menu_df = yelp_menu_df.reset_index(drop=True)

print(yelp_menu_df.shape)
yelp_menu_df.head()
```

```python
yelp_menu_df['location'].value_counts()
```

```python
# Each entry of the 'menu' column is a list of dictionaries
yelp_menu_df['menu']
```

```python
# Unpack all dictionaries from each list and assign them their own row
yelp_exploded_df = yelp_menu_df.explode('menu', ignore_index=True)

yelp_exploded_df
```

```python
# Extract all items from each dictionary
dish_names = yelp_exploded_df['menu'].apply(lambda x: x.get('item_name'))
dish_descriptions = yelp_exploded_df['menu'].apply(lambda x: x.get('description'))
dish_prices = yelp_exploded_df['menu'].apply(lambda x: x.get('price'))

# Create new columns
yelp_exploded_df['menu_item'] = dish_names + ': ' + dish_descriptions # Combine dish names and their descriptions
yelp_exploded_df['item_price'] = dish_prices
```
